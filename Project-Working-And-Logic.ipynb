```python
data_conc_json = pd.DataFrame()
for INP_DATE in date_string:
    data = requests.get("https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/calendarByDistrict?district_id={}&date={}".format(199, INP_DATE),headers = browser_header)
    if (data.ok) and ('centers' in data.json()):
        data_conc_json = data.json()['centers']
        if data_conc_json is not None:
            data_conc_json = pd.DataFrame(data_conc_json)
            if(len(data_conc_json)):
                data_conc_json = data_conc_json.explode('sessions')
                data_conc_json['date']= data_conc_json.sessions.apply(lambda x: x['date'])
                data_conc_json['available_capacity']= data_conc_json.sessions.apply(lambda x: x['available_capacity'])
                data_conc_json['min_age_limit']= data_conc_json.sessions.apply(lambda x: x['min_age_limit'])
                data_conc_json['vaccine']= data_conc_json.sessions.apply(lambda x: x['vaccine'])
                data_conc_json = data_conc_json[["date","state_name", "district_name", "name", "block_name", "pincode", "available_capacity", "min_age_limit", "vaccine", "fee_type"]]
                break
data_conc_json
```
